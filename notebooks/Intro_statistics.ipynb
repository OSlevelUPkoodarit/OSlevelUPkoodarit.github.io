###### Normaalijakauma

Mikli muuttujan arvot ovat normaalijakautuneita, niin tllin suurin osa havaintoja on lhell keskiarvoja, ja mit kauempana keskiarvosta tulos on, sit eptodennkisemp on muuttujan arvojen sijoittuminen e.m. "paikkaan"

Normaalijakaumaa noudattelevia muuttujia: 

- Ihmisten pituudet
- Mittausvirheet
- Koetulokset
- Verenpaine
- Ihmisten lykkyysosamr
 ym. 
 

Normaalijakaumaan liittyy kaksi olennaista arvoa, keskiarvo ja keskihajonta. Keskiarvo kuvaa jakauman huipun arvoa. Keskihajonnalla taas tarkoitetaan sit, miten leve kyr on. Keskihajonnan pienentyess kyr mys kapenee. Noin kaksi kolmasosaa kaikista arvoista mahtuu keskihajonnan phn keskiarvosta. Normitetussa normaalijakaumassa keskiarvo on 0 ja keskihajonta on 1. Tllaisessa tapauksessa siis kaksi kolmasosaa arvoista olisi -1 ja 1 vlill.
