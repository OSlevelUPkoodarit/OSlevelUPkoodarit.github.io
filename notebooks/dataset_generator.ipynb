```python
import pandas as pd
import numpy as np
import random

dummy_firstnames=['Anni', 'Kalle', 'Joonas', 'Pertti', 'Liisa', 'Sanna', 'Outi', 'Charles', 'Dan', 'Emily', 'Dina', 'Carlos']
dummy_lastnames=['Virtanen', 'Kallio', 'Nieminen', 'Korhonen', 'MÃ¤kinen', 'Smith', 'Doe', 'Sanchez', 'Atkins', 'Ruiz', 'Laine']
dummy_gender=['Female', 'Male', 'Other']

indeces=['Student_id', 'First_name', 'Last_name', 'Age', 'Gender', 'Points', 'Enrolled']

first_names=np.random.choice(dummy_firstnames, 1000)
last_names=np.random.choice(dummy_lastnames, 1000)
genders=np.random.choice(dummy_gender, 1000)

student_ids=range(10000, 11000)
mean=34
stdev=18/3

values = [random.gauss(mean, stdev) for _ in range(1000)]

ages=np.floor(values)
points=np.random.randint(0, 301, 1000)
enrolled=np.random.randint(0, 2, 1000)
```

```python
## Construct rows and add to array
data=[]
for i in range(0,1000):
    data.append(pd.Series(data=[student_ids[i], first_names[i], last_names[i], ages[i], genders[i], points[i], enrolled[i]], index=indeces))
    
## Construct dataframe of the array
df=pd.DataFrame(data)

df.head(10)
    
```

```python
df['Student_id']=df['Student_id'].astype('int')
df['Age']=df['Age'].astype('int')
df['Points']=df['Points'].astype('int')
df['Enrolled']=df['Enrolled'].astype('int')

mat = df[['Age','Gender']].values
df_without=df.drop(['Age', 'Gender'], axis=1)

prop = int(mat.size * 0.05)

i = [np.random.choice(range(mat.shape[0])) for _ in range(prop)]
j = [np.random.choice(range(mat.shape[1])) for _ in range(prop)]

mat[i,j] = np.NaN

df=pd.DataFrame(data=mat, columns=['Age', 'Gender'])
df=df_without.join(df)

df.head()
```
